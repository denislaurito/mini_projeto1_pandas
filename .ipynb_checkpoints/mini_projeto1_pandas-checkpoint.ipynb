### Item 7
- Criar coluna "Total" somando o valor de todos os ativos por investidor

```python
df_novoInvestUnidos["Total"] = df_novoInvestUnidos["Ativo01"] + df_novoInvestUnidos["Ativo02"] + df_novoInvestUnidos["Ativo03"] + df_novoInvestUnidos["Ativo04"] + df_novoInvestUnidos["Ativo05"] + df_novoInvestUnidos["Ativo06"] + df_novoInvestUnidos["Ativo07"]
df_novoInvestUnidos
```

- Criar coluna "Media" com a média das notas 1 e 2

```python
df_novoInvestUnidos["Media"] =  (df_novoInvestUnidos["Nota1"] + df_novoInvestUnidos["Nota2"]) / 2
df_novoInvestUnidos
```

- Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)

```python
dom = df_novoInvestUnidos["Email"].str.split("@",n=1, expand = True)
df_novoInvestUnidos["Dominio"] = dom[1]
df_novoInvestUnidos
```

- Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)

```python
df_novoInvestUnidos.rename(columns= {"Salario": "Renda"}, inplace = True)
cifrao = df_novoInvestUnidos["Renda"].str.split("$",n=1, expand = True)
df_novoInvestUnidos["Renda"] = cifrao[1]
df_novoInvestUnidos
```

- Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar a função map)

```python
mudar = df_novoInvestUnidos["Sexo"].map({"Masc" : "Masculino" , "Fem" : "Feminino"})
df_novoInvestUnidos["Sexo"] = mudar
df_novoInvestUnidos
```

- Transformar os dados da coluna Media em inteiro

```python
inteiro = df_novoInvestUnidos["Media"].astype(int)
df_novoInvestUnidos["Media"] = inteiro
df_novoInvestUnidos
```

```python
df_novoInvestUnidos.info()
```

- Criar coluna "NPS" onde o investidor é classificado como promotor se a média das notas é maior ou igual a 80, como neutro se a média estiver entre 50 e 80, e como detrator se for abaixo de 50

```python
df_novoInvestUnidos.loc[df_novoInvestUnidos['Media'] >= 80, 'NPS'] = 'promotor'
df_novoInvestUnidos.loc[(df_novoInvestUnidos['Media'] >= 50) & (df_novoInvestUnidos['Media'] < 80), 'NPS'] = 'neutro'
df_novoInvestUnidos.loc[df_novoInvestUnidos['Media'] < 50, 'NPS'] = 'detrator'
df_novoInvestUnidos
```

### Item 8
- Calcule o total investido em cada coluna Ativo (tente usar o apply)

```python
def somar(a):
    return sum(a)
df_novoInvestUnidos[["Ativo01", "Ativo02", "Ativo03", "Ativo04", "Ativo05", "Ativo06", "Ativo07"]].apply(somar)
```

- Qual o menor valor investido em cada ativo?

```python
def minimo(a):
    return min(a)
df_novoInvestUnidos[["Ativo01", "Ativo02", "Ativo03", "Ativo04", "Ativo05", "Ativo06", "Ativo07"]].apply(minimo)
```

- Mostre a média e desvio padrão de cada ativo

```python
import numpy as np
def media(a):
    return np.average(a)
def desvio(a):
    return np.std(a)
print(f'Media\n{df_novoInvestUnidos[["Ativo01", "Ativo02", "Ativo03", "Ativo04", "Ativo05", "Ativo06", "Ativo07"]].apply(media)}')
print(f'\nDesvio Padrão\n{df_novoInvestUnidos[["Ativo01", "Ativo02", "Ativo03", "Ativo04", "Ativo05", "Ativo06", "Ativo07"]].apply(desvio)}')
```
